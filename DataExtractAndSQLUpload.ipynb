# **Importhig required Python libraries**

```python
import os
import json
import pandas as pd
from sqlalchemy import create_engine
import mysql.connector as sql
```

# **Extracting Data from JSON filre to Data Frame in Pandas**

```python
# Extracting Aggregated Transaction Details

path_1 ="C:\\Users\\LENOVO\\Desktop\\Guvi Proj\\Phonepe\\pulse-master\\data\\aggregated\\transaction\country\\india\\state\\"
Agg_tran_state_list = os.listdir(path_1)

Agg_tra = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "TransactionType":[],
           "TransactionCount":[],
           "TransactionAmount":[]}


for state in Agg_tran_state_list:
    current_state = path_1 + state + "/"
    Agg_yr = os.listdir(current_state)

    for year in Agg_yr:
        current_year = current_state + year + "/"
        Agg_yr_list = os.listdir(current_year)

        for file in Agg_yr_list:
            current_file = current_year + file
            data = open(current_file,"r")
            A= json.load(data)


            for i in A["data"]["transactionData"]: 
   
                Agg_tra["States"].append(state)
                Agg_tra["Years"].append(year)
                Agg_tra["Quarter"].append(int(file.strip(".json")))

                name = i["name"]
                Agg_tra["TransactionType"].append(name)

                count = i["paymentInstruments"][0]["count"]
                Agg_tra["TransactionCount"].append(count)

                amount = i["paymentInstruments"][0]["amount"]
                Agg_tra["TransactionAmount"].append(amount)

df_aggregated_transaction = pd.DataFrame(Agg_tra)
```

```python
# Extracting Aggregated User

path_2 ="C:\\Users\\LENOVO\\Desktop\\Guvi Proj\\Phonepe\\pulse-master\\data\\aggregated\\user\\country\\india\\state\\"
Agg_user_state_list = os.listdir(path_2)

Agg_user = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "Brands":[],
           "TransactionCount":[],
           "Percentage":[]}


for state in Agg_user_state_list:
    current_state = path_2 + state + "/"
    Agg_yr = os.listdir(current_state)

    for year in Agg_yr:
        current_year = current_state + year + "/"
        Agg_yr_list = os.listdir(current_year)

        for file in Agg_yr_list:
            current_file = current_year + file
            data = open(current_file,"r")
            B= json.load(data)
            try:
                for i in B["data"]["usersByDevice"]: 
    
                    Agg_user["States"].append(state)
                    Agg_user["Years"].append(year)
                    Agg_user["Quarter"].append(int(file.strip(".json")))

                    brand = i["brand"]
                    Agg_user["Brands"].append(brand)

                    count = i["count"]
                    Agg_user["TransactionCount"].append(count)

                    percentage = i["percentage"]
                    Agg_user["Percentage"].append(percentage)
            
            except : 
                pass


df_aggregated_user = pd.DataFrame(Agg_user)
```

```python
# Extracting Map Transaction Details

path_3 ="C:\\Users\\LENOVO\\Desktop\\Guvi Proj\\Phonepe\\pulse-master\\data\\map\\transaction\\hover\\country\\india\\state\\"
map_tra_state_list = os.listdir(path_3)

map_tra = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "Districts":[],
           "TransactionCount":[],
           "TransactionAmount":[]}


for state in map_tra_state_list:
    current_state = path_3 + state + "/"
    Agg_yr = os.listdir(current_state)

    for year in Agg_yr:
        current_year = current_state + year + "/"
        Agg_yr_list = os.listdir(current_year)

        for file in Agg_yr_list:
            current_file = current_year + file
            data = open(current_file,"r")
            C= json.load(data)


            for i in C["data"]["hoverDataList"]: 
    
                    map_tra["States"].append(state)
                    map_tra["Years"].append(year)
                    map_tra["Quarter"].append(int(file.strip(".json")))

                    name = i["name"]
                    map_tra["Districts"].append(name)

                    count = i["metric"][0]["count"]
                    map_tra["TransactionCount"].append(count)

                    amount = i["metric"][0]["amount"]
                    map_tra["TransactionAmount"].append(amount)
            
df_map_transaction = pd.DataFrame(map_tra)
```

```python
# Extracting Map User Details

path_4 ="C:\\Users\\LENOVO\\Desktop\\Guvi Proj\\Phonepe\\pulse-master\\data\\map\\user\\hover\\country\\india\\state\\"
map_user_state_list = os.listdir(path_4)

map_user = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "Districts":[],
           "RegisteredUsers":[],
           "AppOpens":[]}


for state in map_user_state_list:
    current_state = path_4 + state + "/"
    Agg_yr = os.listdir(current_state)

    for year in Agg_yr:
        current_year = current_state + year + "/"
        Agg_yr_list = os.listdir(current_year)

        for file in Agg_yr_list:
            current_file = current_year + file
            data = open(current_file,"r")
            D= json.load(data)


            for i in D["data"]["hoverData"].items(): 
    
                    map_user["States"].append(state)
                    map_user["Years"].append(year)
                    map_user["Quarter"].append(int(file.strip(".json")))

                    name = i[0]
                    map_user["Districts"].append(name)

                    registeredUsers = i[1]["registeredUsers"]
                    map_user["RegisteredUsers"].append(registeredUsers)

                    appOpens = i[1]["appOpens"]
                    map_user["AppOpens"].append(appOpens)
            
df_map_user = pd.DataFrame(map_user)
```

```python
# Extracting Top Transaction Details

path_5 ="C:\\Users\\LENOVO\\Desktop\\Guvi Proj\\Phonepe\\pulse-master\\data\\top\\transaction\\country\\india\\state\\"
top_tra_state_list = os.listdir(path_5)


top_tra = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "Pincodes":[],
           "TransactionCount":[],
           "TransactionAmount":[]}


for state in top_tra_state_list:
    current_state = path_5 + state + "/"
    Agg_yr = os.listdir(current_state)

    for year in Agg_yr:
        current_year = current_state + year + "/"
        Agg_yr_list = os.listdir(current_year)

        for file in Agg_yr_list:
            current_file = current_year + file
            data = open(current_file,"r")
            E= json.load(data)


            for i in E["data"]["pincodes"]:    
               
                top_tra["States"].append(state)
                top_tra["Years"].append(year)
                top_tra["Quarter"].append(int(file.strip(".json")))

                entityName = i["entityName"]
                top_tra["Pincodes"].append(entityName)

                count = i["metric"]["count"]
                top_tra["TransactionCount"].append(count)

                amount = i["metric"]["amount"]
                top_tra["TransactionAmount"].append(amount)
                
                
            
df_top_transaction = pd.DataFrame(top_tra)
```

```python
# Extracting Top User Details

path_6 ="C:\\Users\\LENOVO\\Desktop\\Guvi Proj\\Phonepe\\pulse-master\\data\\top\\user\\country\\india\\state\\"
top_user_state_list = os.listdir(path_6)

top_user = {"States":[], 
           "Years":[],
           "Quarter" :[],
           "Pincodes":[],
           "RegisteredUsers":[]}


for state in top_user_state_list:
    current_state = path_6 + state + "/"
    Agg_yr = os.listdir(current_state)

    for year in Agg_yr:
        current_year = current_state + year + "/"
        Agg_yr_list = os.listdir(current_year)

        for file in Agg_yr_list:
            current_file = current_year + file
            data = open(current_file,"r")
            F= json.load(data)


            for i in F["data"]["pincodes"]:    
               
                top_user["States"].append(state)
                top_user["Years"].append(year)
                top_user["Quarter"].append(int(file.strip(".json")))

                entityName = i["name"]
                top_user["Pincodes"].append(entityName)

                registeredusers = i["registeredUsers"]
                top_user["RegisteredUsers"].append(registeredusers)


df_top_user = pd.DataFrame(top_user)
```

**Converting and Downloading all the Data Frame files to CSV**

```python
df_aggregated_transaction.to_csv('df_aggregated_transaction.csv')
df_aggregated_user.to_csv('df_aggregated_user.csv')
df_map_transaction.to_csv('df_map_transaction.csv')
df_map_user.to_csv('df_map_user.csv')
df_top_transaction.to_csv('df_top_transaction.csv')
df_top_user.to_csv('df_top_user.csv')
```

```python
# SQL CONNECTION:
mydb = sql.connect(
    host="localhost",
    user="root",
    password="root"
)
# Create a cursor object using the connection
mycursor = mydb.cursor()

# Create Database if not exists
mycursor.execute("CREATE DATABASE IF NOT EXISTS phonepe")

# Use the specific database
mycursor.execute("USE phonepe")

#Create df_aggregated_transaction
mycursor.execute("CREATE TABLE IF NOT EXISTS df_aggregated_transaction (\
                    States VARCHAR(255),\
                    Years VARCHAR(255),\
                    Quarter INT,\
                    TransactionType VARCHAR(255),\
                    TransactionCount INT,\
                    TransactionAmount FLOAT)")
mydb.commit()

#Create df_aggregated_user
mycursor.execute("CREATE TABLE IF NOT EXISTS df_aggregated_user(\
                    States VARCHAR(255),\
                    Years VARCHAR(255),\
                    Quarter INT,\
                    Brands VARCHAR(255),\
                    TransactionCount INT,\
                    Percentage FLOAT)")
mydb.commit()

#Create df_map_transaction
mycursor.execute("CREATE TABLE IF NOT EXISTS df_map_transaction (\
                    States VARCHAR(255),\
                    Years VARCHAR(255),\
                    Quarter INT,\
                    Districts VARCHAR(255),\
                    TransactionCount INT,\
                    TransactionAmount FLOAT)")
mydb.commit()

#Create df_map_user
mycursor.execute("CREATE TABLE IF NOT EXISTS df_map_user (\
                    States VARCHAR(255),\
                    Years VARCHAR(255),\
                    Quarter INT,\
                    Districts VARCHAR(255),\
                    RegisteredUsers INT,\
                    AppOpens INT)")
mydb.commit()

#Create df_top_transaction
mycursor.execute("CREATE TABLE IF NOT EXISTS df_top_transaction (\
                    States VARCHAR(255),\
                    Years VARCHAR(255),\
                    Quarter INT,\
                    Pincodes VARCHAR(255),\
                    TransactionCount INT,\
                    TransactionAmount FLOAT)")
mydb.commit()

#Create df_top_user
mycursor.execute("CREATE TABLE IF NOT EXISTS df_top_user (\
                    States VARCHAR(255),\
                    Years VARCHAR(255),\
                    Quarter INT,\
                    Pincodes VARCHAR(255),\
                    RegisteredUsers INT)")
mydb.commit()
```

```python

## Upload Data on SQL

engine = create_engine("mysql+mysqlconnector://root:root@localhost/phonepe")

df_aggregated_transaction.to_sql(name="df_aggregated_transaction", con=engine, if_exists='append', index=False)
df_aggregated_user.to_sql(name="df_aggregated_user", con=engine, if_exists='append', index=False)
df_map_transaction.to_sql(name="df_map_transaction", con=engine, if_exists='append', index=False)
df_map_user.to_sql(name="df_map_user", con=engine, if_exists='append', index=False)
df_top_transaction.to_sql(name="df_top_transaction", con=engine, if_exists='append', index=False)
df_top_user.to_sql(name="df_top_user", con=engine, if_exists='append', index=False)
```
